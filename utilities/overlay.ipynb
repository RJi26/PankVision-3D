```python
img = 'pancreas_409_0.nii.gz'
path_i = os.path.join(path_img, img)
final_img = nib.load(path_i)
data1 = final_img.get_fdata()

# Load the second nifti image
path_l = os.path.join(path_lab, img)
final_lab = nib.load(path_l)
data2 = final_lab.get_fdata()
# Create a custom colormap
%matplotlib inline
my_cmap, _ = from_levels_and_colors(levels=[0, 1, 2], colors=[[0., 0., 0., 0.], [0., 1., 0., 1], [0.8, 0., 0.8, 1]], extend='max')

def plot_slice(slice_number):
    plt.figure(figsize=(8, 8))
    plt.imshow(data1[:, :, slice_number], cmap='gray')
    plt.imshow(data2[:, :, slice_number].astype('int'), cmap=my_cmap, vmin=0., vmax=1.)
    plt.title(f"Slice {slice_number + 1}")
    plt.axis('off')
    plt.show()

# Create the interactive slider widget
slice_slider = widgets.IntSlider(value=0, min=0, max=data1.shape[2] - 1, step=1, description='Slice:')
widgets.interactive(plot_slice, slice_number=slice_slider)
```
