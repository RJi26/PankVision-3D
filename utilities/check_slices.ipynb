```python
import nibabel as nib

def count_dicom_slices(nifti_filename):
    try:
        nifti_img = nib.load(nifti_filename)
        num_slices = nifti_img.shape[2]  # Assuming slices are in the third dimension
        return num_slices
    except Exception as e:
        return f"Error: {e}"

if __name__ == "__main__":
    file_name = 'label0003.nii.gz'
    path_labels = '/content/drive/MyDrive/Machine-Learning-Biomedicine/Pancreatic-Cancer/model3-TCIA/data/labels/'
    nifti_file = os.path.join(path_labels, file_name)
    num_slices = count_dicom_slices(nifti_file)

    if isinstance(num_slices, int):
        print(f"The NIfTI file '{nifti_file}' contains {num_slices} slices.")
    else:
        print(num_slices)
```
