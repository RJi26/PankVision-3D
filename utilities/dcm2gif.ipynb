```python
import nibabel as nib
import imageio
import os

# Load the NIfTI file
img = nib.load('/content/drive/MyDrive/Machine-Learning-Biomedicine/PankVision-3D/dataset/dataset-007/nifti_files/images/pancreas_001_0.nii.gz')

# Get the data array from the image
data = img.get_fdata()

# Create a list to hold each slice as an image
images = []

# Loop through each slice in the third dimension
for i in range(data.shape[2]):
    # Append the slice to the images list
    images.append(data[:, :, i])

# Specify your output directory
output_dir = '/content/drive/MyDrive/Machine-Learning-Biomedicine/PankVision-3D/img'

# Make sure the directory exists, if not, create it
os.makedirs(output_dir, exist_ok=True)

# Save the slices as a GIF to the specified directory
imageio.mimsave(os.path.join(output_dir, 'output.gif'), images, 'GIF', duration=0.1)
```
